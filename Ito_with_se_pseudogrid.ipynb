```python
port_number = 12346  # Puedes ajustar este número de puerto si es necesario
NUM_GPUs = 1  # Google Colab generalmente tiene una sola GPU
checkpoint_path = '/content/log/ITO/adaptiveweight_dp_fc1_avg_1727307290/best.pth'
cfg_path = '/content/se-pseudogrid/cfgs/datasetITO/adaptiveweight_dp_fc1_avg.yaml'
data_aug = True  # Ajusta según tus necesidades
LOSS = 'smooth'  # Ajusta según tus necesidades
use_avg_max_pool = True  # Ajusta según tus necesidades
```

```python
# Crear un archivo run_torchrun.py
with open('run_torchrun.py', 'w') as f:
    f.write(f'''
import os
import subprocess
import torch.distributed as dist

# Configurar las variables necesarias
port_number = {port_number}  # Puedes ajustar este número de puerto si es necesario
NUM_GPUs = {NUM_GPUs}  # Google Colab generalmente tiene una sola GPU
checkpoint_path = '{checkpoint_path}'
cfg_path = '{cfg_path}'
data_aug = {data_aug}  # Ajusta según tus necesidades
LOSS = '{LOSS}'  # Ajusta según tus necesidades
use_avg_max_pool = {use_avg_max_pool}  # Ajusta según tus necesidades

def main():
    # Establecer LOCAL_RANK a 1 por defecto
    os.environ['LOCAL_RANK'] = '1'

    command = [
        '/usr/local/bin/torchrun',
        '--nproc_per_node=' + str(NUM_GPUs),
        '--master_port=' + str(port_number),
        'se-pseudogrid/function/evaluate_ITO_dist.py',
        '--cfg', cfg_path,
        '--load_path', checkpoint_path,
        '--data_aug', str(data_aug),
        '--loss', LOSS,
        '--use_avg_max_pool', str(use_avg_max_pool)
    ]

    # Ejecutar el comando
    result = subprocess.run(command, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

    # Imprimir la salida y el error
    print(result.stdout.decode('utf-8'))
    print(result.stderr.decode('utf-8'))

    # Destruir el grupo de procesos si está inicializado
    if dist.is_initialized():
        dist.destroy_process_group()

if __name__ == "__main__":
    main()
''')

# Ejecutar el script de Python
!python run_torchrun.py
```
